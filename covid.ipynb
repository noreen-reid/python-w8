```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\disma\OneDrive\Desktop\Covid-19\owid-covid-data.csv"
df = pd.read_csv(file_path)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
countries = ['Kenya', 'United States', 'India']
df_countries = df[df['location'].isin(countries)]
```

```python
columns = ['location', 'date', 'total_cases', 'total_deaths', 
           'total_vaccinations', 'people_vaccinated', 
           'people_fully_vaccinated', 'population']
df_countries = df_countries[columns]
```

```python
df_clean = df_countries.dropna(subset=['total_cases', 'total_deaths'])
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_clean, x='date', y='total_cases', hue='location')
plt.title('Total COVID-19 Cases Over Time')
...
plt.show()
```

```python
df_vax = df_countries.dropna(subset=['total_vaccinations'])

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_vax, x='date', y='total_vaccinations', hue='location')
plt.title('Total COVID-19 Vaccinations Over Time')
...
plt.show()
```

```python
df_vax['percent_vaccinated'] = df_vax['people_vaccinated'] / df_vax['population'] * 100
```

```python
df_vax.loc[:, 'percent_vaccinated'] = df_vax['people_vaccinated'] / df_vax['population'] * 100
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_vax, x='date', y='percent_vaccinated', hue='location', marker='o')
plt.title('COVID-19 Vaccination Coverage (% of Population)')
plt.xlabel('Date')
plt.ylabel('% Vaccinated (at least one dose)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##Key Insights
The United States had the fastest and most aggressive vaccine rollout, quickly reaching high vaccination coverage by mid-2021.

India's vaccination campaign accelerated sharply after initial delays, overtaking the U.S. in total doses administered due to its larger population.

Kenya experienced a slower vaccine distribution, with a significantly lower percentage of its population vaccinated during the same time period.

COVID-19 cases and deaths in the United States remained consistently higher than in India and Kenya, reflecting both population size and early widespread transmission.

India showed a steep spike in cases and deaths around mid-2021, coinciding with the Delta variant wave.

 ##Notable Patterns and Anomalies
Kenya's data has more missing values, especially early in the pandemic, likely due to limited testing/reporting capacity.

Despite having more total vaccinations, India’s per-capita vaccination rate lagged early on, highlighting the scale of its public health challenge.

The gap between people vaccinated and fully vaccinated was more visible in the early months, particularly in countries using two-dose regimens.

# Narrative Summary
This analysis reveals the disparities in pandemic response and vaccination efforts across nations. While the U.S. had better initial access to vaccines and health infrastructure, India made rapid progress through mass immunization drives. Kenya’s slower uptake underscores challenges in vaccine availability and distribution in lower-income countries. Such insights are critical for informing future public health strategies
